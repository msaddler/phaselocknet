```python
"""
Load all word/voice recognition experiment data.
Compute min/max proportion of words and voices correct for humans across
different conditions (used to normalize RMSE metric calculations).
"""

fn_data = "data/model/cochlearn_spkr_word_recognition.pkl"
with open(fn_data, "rb") as f:
    EXPERIMENT_DATAFRAMES = pickle.load(f)
print(f"Loaded: {fn_data}")
for tag_expt in EXPERIMENT_DATAFRAMES.keys():
    print(f"|__ {tag_expt}")

min_human_word_correct = np.inf
max_human_word_correct = -np.inf
min_human_spkr_correct = np.inf
max_human_spkr_correct = -np.inf
for tag_expt in ["kell_like", "pitch_altered", "speech_in_synthetic_textures"]:
    df = EXPERIMENT_DATAFRAMES[tag_expt]
    df = df[df.tag_model == "human"]
    min_h_word, max_h_word = df.correct_word_mean.min(), df.correct_word_mean.max()
    min_h_spkr, max_h_spkr = df.correct_spkr_mean.min(), df.correct_spkr_mean.max()
    min_human_word_correct = min(min_human_word_correct, min_h_word)
    max_human_word_correct = max(max_human_word_correct, max_h_word)
    min_human_spkr_correct = min(min_human_spkr_correct, min_h_spkr)
    max_human_spkr_correct = max(max_human_spkr_correct, max_h_spkr)
# print(f"{min_human_word_correct=}")
# print(f"{max_human_word_correct=}")
# print(f"{min_human_spkr_correct=}")
# print(f"{max_human_spkr_correct=}")
```

```python
"""
Calculate human-model comparison metrics for
all word/voice recognition experiments and save
output to a file.
"""

dict_metric_function = {
    "pearsonr": scipy.stats.pearsonr,
    "rmse": util.normalized_rmse,
}

experiment_to_compfunc_map = {
    "kell_like_word": functools.partial(
        util_spkr_word_psychophysics.compare_word_recognition_kell_like,
        restrict_conditions=[0, 1, 2, 3]),
    "speech_in_synthetic_textures": functools.partial(
        util_spkr_word_psychophysics.compare_word_recognition_speech_in_synthetic_textures),
    "pitch_altered_word": functools.partial(
        util_spkr_word_psychophysics.compare_recognition_pitch_altered, key_task="word"),
    "pitch_altered_spkr": functools.partial(
        util_spkr_word_psychophysics.compare_recognition_pitch_altered, key_task="spkr"),
    "hopkins_moore_2009_word": functools.partial(
        util_spkr_word_psychophysics.compare_word_recognition_hopkins_moore_2009),
}

def func_to_parallelize(tag_expt):
    compfunc = experiment_to_compfunc_map[tag_expt]
    tag_expt_for_dataframe = tag_expt
    for suffix in ["_spkr", "_word"]:
        tag_expt_for_dataframe = tag_expt_for_dataframe.replace(suffix, "")
    df = EXPERIMENT_DATAFRAMES[tag_expt_for_dataframe]
    df_results = None
    for metric_key, metric_function_tmp in dict_metric_function.items():
        if metric_key == "pearsonr":
            metric_function = metric_function_tmp
        else:
            if tag_expt in ["kell_like_word", "speech_in_synthetic_textures", "pitch_altered_word"]:
                # Word recognition experiments measuring proportion correct receive same min/max normalization
                metric_function = lambda y_human, y_model: metric_function_tmp(
                    y_human,
                    y_model,
                    min_human=min_human_word_correct,
                    max_human=max_human_word_correct,
                )
            elif tag_expt in ["pitch_altered_spkr"]:
                # Voice recognition experiments measuring proportion correct receive same min/max normalization
                metric_function = lambda y_human, y_model: metric_function_tmp(
                    y_human,
                    y_model,
                    min_human=min_human_spkr_correct,
                    max_human=max_human_spkr_correct,
                )
            else:
                # All other experiments are normalized within experiment (different units and stimuli)
                metric_function = metric_function_tmp
        df_results_tmp = compfunc(
            df,
            bootstrap_repeats=1000,
            metric_function=metric_function)
        df_results_tmp = df_results_tmp.rename(columns={
            "metric": metric_key,
            "list_metric": f"list_{metric_key}",
            "bootstrap_list_metric": f"bootstrap_list_{metric_key}",
        })
        df_results_tmp["tag_expt"] = tag_expt
        if df_results is None:
            df_results = df_results_tmp
        else:
            df_results = df_results.merge(df_results_tmp, on=["tag_expt", "tag_model"])
    return df_results


with multiprocessing.Pool(60) as pool:
    list_df_results = pool.map(func_to_parallelize, list(experiment_to_compfunc_map.keys()))
df_results = pd.concat(list_df_results)

# Set metrics to NaN for rows where human data is unavailable (all zeros give undefined pearsonr)
IDX = df_results["pearsonr"].isna()
df_results.loc[IDX, "rmse"] = df_results.loc[IDX, "pearsonr"]
df_results.loc[IDX, "list_rmse"] = df_results.loc[IDX, "list_pearsonr"]
df_results.loc[IDX, "bootstrap_list_rmse"] = df_results.loc[IDX, "bootstrap_list_pearsonr"]

fn_metrics = "data/model/cochlearn_spkr_word_recognition_human_model_comparison_metrics.pkl"
df_results.to_pickle(fn_metrics)
print(f"Wrote: {fn_metrics}")
```

```python
def make_plot(
    list_list_model=[],
    list_str_title=[],
    list_key_task="localization",
    list_key_metric="performance",
):
    """ """
    if not isinstance(list_list_model[0], list):
        list_list_model = [list_list_model]
        list_str_title = [list_str_title]
    if not isinstance(list_key_task, list):
        list_key_task = [list_key_task] * len(list_list_model)
    if not isinstance(list_key_metric, list):
        list_key_metric = [list_key_metric] * len(list_list_model)
    gridspec_kw = {
        "width_ratios": [len(_) for _ in list_list_model],
    }
    fig, ax_arr = plt.subplots(
        nrows=1,
        ncols=len(list_list_model),
        figsize=(3.5 * len(list_list_model), 4),
        gridspec_kw=gridspec_kw,
    )
    ax_arr = np.array(ax_arr).reshape([len(list_list_model)])
    for itr_ax in range(len(list_list_model)):
        ax = ax_arr[itr_ax]
        list_model = list_list_model[itr_ax]
        key_task = list_key_task[itr_ax]
        key_metric = list_key_metric[itr_ax]
        list_tag_model = [dict_tag_model[key_task][_] for _ in list_model]
        kwargs = {
            "fn_data_localization": "data/model/cochlearn_sound_localization.pkl",
            "fn_data_spkr_word": "data/model/cochlearn_spkr_word_recognition.pkl",
            "fn_comparison_localization": "data/model/cochlearn_sound_localization_human_model_comparison_metrics.pkl",
            "fn_comparison_spkr_word": "data/model/cochlearn_spkr_word_recognition_human_model_comparison_metrics.pkl",
            "list_tag_model": list_tag_model,
            "key_task": key_task,
            "key_metric": key_metric,
        }
        y, list_y, bootstrap_list_y = util.get_aggregate_measure(**kwargs)
        list_x = []
        xticks = []
        xticklabels = []
        if key_metric == "performance":
            print(f"{key_task=}")
            for k, v in zip(list_model, y):
                print("|___ {} ({}): {:.1f}".format(key_metric, k, v))
        for x, tag_model in enumerate(list_tag_model):
            color, label = util.get_color_and_label_from_model_tag(tag_model)
            facecolor = color
            label = label.replace(" Hz IHC filter", "")
            if "cochlearn" in tag_model:
                color = "green"
                facecolor = "green"
                label = "learn"
            if "delayed" in tag_model:
                facecolor = "orange"
                label = label + "\n" + r"$^{\text{(delayed)}}$"
            parts = ax.violinplot(
                bootstrap_list_y[x],
                positions=[x],
                showmeans=False,
                showextrema=False)
            for k in parts.keys():
                if not k == "bodies":
                    parts[k].set_color(color)
                    parts[k].set_linewidth(2)
            for pc in parts["bodies"]:
                pc.set_facecolor(facecolor)
                pc.set_edgecolor(color)
                pc.set_linewidth(1.5)
                pc.set_alpha(0.75)
            ax.plot(
                x,
                y[x],
                color=color,
                marker="o",
                ms=4,
                mew=1.5,
                label=label,
                mfc=facecolor)
            list_x.append(x)
            xticks.append(x)
            xticklabels.append(label)
        kwargs_plot = {
            "color": "k",
            "lw": 2,
            "ls": ":",
            "marker": "",
            "zorder": -100,
        }
        ax.plot(list_x, y, **kwargs_plot)
        dict_kwargs_format_axes = {
            "pearsonr": {
                "localization": {"ylimits": [0.5, 1.0], "str_ylabel": "\nHuman-model similarity (Pearson $r$)"},
                "spkr": {"ylimits": [0.5, 1.0], "str_ylabel": "\nHuman-model similarity (Pearson $r$)"},
                "word": {"ylimits": [0.5, 1.0], "str_ylabel": "\nHuman-model similarity (Pearson $r$)"},
            },
            "rmse": {
                "localization": {"ylimits": [2, 0], "yticks": [2.0, 1.5, 1.0, 0.5, 0], "str_ylabel": "\nHuman-model similarity (RMS error)"},
                "spkr": {"ylimits": [2, 0], "yticks": [2.0, 1.5, 1.0, 0.5, 0], "str_ylabel": "\nHuman-model similarity (RMS error)"},
                "word": {"ylimits": [2, 0], "yticks": [2.0, 1.5, 1.0, 0.5, 0], "str_ylabel": "\nHuman-model similarity (RMS error)"},
            },
            "performance": {
                "localization": {"ylimits": [50, 0], "str_ylabel": "\nSound localization (° absolute error)"},
                "spkr": {"ylimits": [50, 100], "str_ylabel": "\nVoice recognition (% correct)"},
                "word": {"ylimits": [25, 75], "yticks": [25, 35, 45, 55, 65, 75], "str_ylabel": "\nWord recognition (% correct)"},
            }
        }
        ax = util_figures.format_axes(
            ax,
            xticks=xticks,
            xticklabels=xticklabels,
            str_xlabel="Model IHC filter cutoff (Hz)" if itr_ax == 0 else None,
            str_title=list_str_title[itr_ax],
            fontweight_title="bold",
            **dict_kwargs_format_axes[key_metric][key_task],
        )
        [min_y, max_y] = dict_kwargs_format_axes[key_metric][key_task]["ylimits"]
        list_y_stat = np.linspace(min_y, max_y, 12)
        for itr, (idx_ref, idx_cmp) in enumerate([(3, 4)]):
            y_stat = list_y_stat[itr + 3]
            y_stat_text = y_stat - 0.8 * (np.diff(list_y_stat)[0])
            x_ref = list_x[idx_ref]
            x_cmp = list_x[idx_cmp]
            y_ref_dist = bootstrap_list_y[idx_ref]
            y_cmp_dist = bootstrap_list_y[idx_cmp]
            y_ref = y[idx_ref]
            y_cmp = y[idx_cmp]
            p_val = util.get_p_val(y_ref_dist, y_cmp)
            d_val = util.cohend(y_ref_dist, y_cmp_dist)
            ax.plot(
                [x_ref, x_cmp],
                [y_stat, y_stat],
                marker=2,
                color="k",
                lw=2,
                mew=2,
                ms=4)
            str_text = "$p<0.001$" if p_val < 0.001 else "$p={:.3f}$".format(p_val)
            ax.text(
                np.mean([x_ref, x_cmp]),
                y_stat_text,
                str_text,
                ha="center",
                va="bottom",
                fontsize=10,
                fontweight="bold",
            )
    return fig, ax_arr


for key_metric in ["performance", "pearsonr", "rmse"]:
    fig, ax_arr = make_plot(
        list_list_model=[
            ["50", "320", "1000", "3000", "cochlearn", "3000_delayed", "cochlearn_delayed"],
            ["50", "320", "1000", "3000", "cochlearn"],
            ["50", "320", "1000", "3000", "cochlearn"],
        ],
        list_str_title=[
            "Sound localization",
            "Voice recognition",
            "Word recognition",
        ],
        list_key_task=["localization", "spkr", "word"],
        list_key_metric=key_metric,
    )
    for itr_ax in range(len(ax_arr)):
        ax_arr[itr_ax].text(
            0.95,
            0.05,
            "GREEN: model with\nlearned cochlea",
            fontsize=10,
            fontweight="bold",
            color="green",
            ha="right",
            va="bottom",
            transform=ax_arr[itr_ax].transAxes,
        )
    plt.tight_layout()
    plt.show()

    fn_fig = os.path.join(f"{dir_fig}/cochlearn_aggregate_{key_metric}.pdf")
    fig.savefig(fn_fig, bbox_inches="tight", pad_inches=0.05, transparent=True)
    print(fn_fig + "\n\n")
```
